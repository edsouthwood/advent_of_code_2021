```python
import numpy as np

# import data and make it into a list of coord pairs
with open('data.txt') as f:
    data = f.readlines()
data = [d.strip('\n') for d in data]
data = [d.split(' -> ') for d in data]
data = [[d[0].split(','),d[1].split(',')] for d in data]

#convert all to ints
for i, pair in enumerate(data):
    #print(pair, i)
    for j, coord in enumerate(pair):
        for k,x in enumerate(coord):
            data[i][j][k] = int(x)

#sort the data points
for i, pair in enumerate(data):
    data[i].sort()

# find the dimensions of the grid notice the +1 because count from zero
def find_max_dimensions(data_list):
    x_max = 0
    y_max = 0
    for pair in data:
        for coord in pair:
            if int(coord[0])>x_max:
                x_max = int(coord[0])
            if int(coord[1])>y_max:
                y_max = int(coord[1])
    return x_max + 1, y_max + 1

# make a grid to plot on
grid = np.zeros(find_max_dimensions(data))

def add_line_to_grid(pair, grid):
    x_1 = int(pair[0][0])
    y_1 = int(pair[0][1])
    x_2 = int(pair[1][0])
    y_2 = int(pair[1][1])
    # check for diagonal lines
    if (x_1 != x_2) and (y_1 != y_2):
        #45 degree lines
        if abs(x_1 - x_2) == abs(y_1 - y_2):
            #print('45 degrees')
            if y_2 > y_1:
                #increasing
                for i in range((y_1-y_1),(y_2-y_1)+1):
                    grid[y_1+i][x_1+i] += 1
            else:
                #decreasing
                #print("drcreasing")
                #print("for points ",x_1,y_1,x_2,y_2)
                #print(list(range(y_1-y_2,(y_2-y_2)-1,-1)))
                for i in range(y_1-y_2,(y_2-y_2)-1,-1):
                    #print(y_2+i,x_1+(y_1-y_2-i))
                    grid[y_2+i][x_1+(y_1-y_2-i)] += 1
            pass
        else:
            #other diagonals
            pass
    else:#must be horizontal or vertical
        # check vertical line
        if x_1 == x_2:
            for y in range(min(y_1,y_2),max(y_1,y_2)+1):
                grid[y][x_1] += 1
        else:# check horizontal lines
            for x in range(min(x_1,x_2),max(x_1,x_2)+1):
                grid[y_1][x] += 1
    return grid

# for each pair of coords add to grid
for pair in data:
    grid = add_line_to_grid(pair,grid)

# cycle through elements in grid and count
def count_multi_coords(grid):
    count = 0
    for row in grid:
        for el in row:
            if el > 1:
                count+=1
    return(count)

ans = count_multi_coords(grid)
print(grid)
print("answer = ",ans)
```
