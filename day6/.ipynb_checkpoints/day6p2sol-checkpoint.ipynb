```python
with open("data.txt") as f:
    data = f.readlines()
data = [d.strip('\n') for d in data]
data = data[0]
data = data.split(',')
data = [int(d) for d in data]
data
fish = [0]*9

for d in data:
    fish[d] +=1
    
def inc_count(data):
    new_fish = [0]*9
    for index, count in enumerate(data):
        #print(index, count)
        if index == 0:
            new_fish[8] += count
            new_fish[6] += count
        else:
            new_fish[index - 1] += count
    return new_fish

for x in range(257):
    print("day ",x, sum(fish))
    fish = inc_count(fish)
#print("day ",x, sum(fish))
```
