```python
# make dictionary of connections
def make_connections(data):
    connections = dict()
    for pair in data:
        if pair[0] in connections.keys():
            connections[pair[0]] += [pair[1]]
        else:
            connections[pair[0]] = [pair[1]]
    for pair in data:
        if pair[1] in connections.keys():
            connections[pair[1]] += [pair[0]]
        else:
            connections[pair[1]] = [pair[0]]
    return connections
```

```python
def make_start_routes(connections):
    start_routes = []
    for connection in connections['start']:
        start_routes += [[connection]]
    return start_routes
```

```python
def has_double(li):
    if li == None:
        return True
    li = [l for l in li if not l.isupper()]
    return not len(li) == len(list(set(li)))
```

```python
def has_double_excluding_x(li,x):
    keep_going = True
    while keep_going:
        try:
            li.remove(x)
        except:
            keep_going = False
    if li == None:
        return True
    li = [l for l in li if not l.isupper()]
    return not len(li) == len(list(set(li)))
```

```python
def get_diff_nodes(data):
    different_nodes = []
    for pair in data:
        for item in pair:
            different_nodes += [item]
    return list(set(different_nodes))
```

```python
def make_one_upper(data, to_try):
    d = data.copy()
    for i, pair in enumerate(d):
        for j, item in enumerate(pair):
            if item == to_try[0]:
                d[i][j] = item.upper()
    return d
```
