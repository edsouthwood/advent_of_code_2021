```python
with open("data.txt") as f:
    data = f.readlines()
data = [d.strip("\n") for d in data]
data = [d.split("-") for d in data]
data
```

```python
# make dictionary of connections
connections = dict()
for pair in data:
    if pair[0] in connections.keys():
        connections[pair[0]] += [pair[1]]
    else:
        connections[pair[0]] = [pair[1]]
for pair in data:
    if pair[1] in connections.keys():
        connections[pair[1]] += [pair[0]]
    else:
        connections[pair[1]] = [pair[0]]
```

```python
def has_double(li):
    if li == None:
        return True
    li = [l for l in li if not l.isupper()]
    return not len(li) == len(list(set(li)))
```
