```python
with open("data.txt") as f:
    data = f.readlines()
data = [d.strip('\n') for d in data]
data = [[d.split(' | ')[0],d.split(' | ')[1]] for d in data]
data
```

```python
count = 0
for d in data:
    for num in d[1].split(" "):
        if len(num) in [2,3,4,7]:
            count +=1
    #print("output ", d[1].split(" "))
print(count)
```
