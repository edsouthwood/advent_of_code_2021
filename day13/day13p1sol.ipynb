```python
points = []
lines = []
with open("data.txt") as f:
    toggle = True
    for line in f:
        if line == '\n':
            toggle = False
        if toggle:
            points += [line.strip('\n')]
        else:
            if line == '\n':
                pass
            else:
                lines += [line.strip('\n')]
print(points)
print(lines)
```

```python
class point():
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def reflect(self, line):
        line = line.split("=")
        if line[0] == 'x':
            if self.x < int(line[1]):
                return
            else:
                self.x = 2*int(line[1]) - self.x
                return
        else:
            if self.y < int(line[1]):
                return
            else:
                self.y = 2*int(line[1]) - self.y
                return
    
    def to_tup(self):
        return tuple([self.x,self.y])
```

```python
pnts = []
for p in points:
    p = p.split(',')
    pnts += [point(int(p[0]),int(p[1]))]
```

```python
lns = []
for l in lines:
    l = l.lstrip("fold along ")
    lns += [l]
```

```python
for p in pnts:
    p.reflect(lns[0])
```

```python
for l in lns:
    for p in pnts:
        #print(p.x,p.y,"reflected in line",l)
        p.reflect(l)
        #print(p.x,p.y)
```
